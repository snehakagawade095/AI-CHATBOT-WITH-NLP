```python
import nltk
from nltk.chat.util import Chat, reflections

pairs = [

    [r"(hi|hello|hey|hii|hola)", 
     ["Hello! ðŸ˜Š How can I help you today?", "Hi there! ðŸ‘‹ What can I do for you?"]],

    [r"what is your name ?",
     ["I am an AI Chatbot created for an internship task."]],

    [r"what can you do ?",
     ["I can answer simple questions using NLP.", 
      "I can chat with you and answer basic queries."]],

    [r"(internship|project|task)",
     ["This chatbot is part of an NLP internship task using Python and NLTK."]],

    [r"(python|programming)",
     ["Python is a powerful and beginner-friendly programming language."]],

    [r"(nlp|natural language processing)",
     ["NLP helps machines understand and respond to human language."]],

    [r"(help|support)",
     ["Sure! Tell me what you need help with."]],

    [r"how are you ?",
     ["I'm doing great! Thanks for asking ðŸ˜Š", "I'm fine! How about you?"]],

    [r"(thanks|thank you)",
     ["You're welcome! ðŸ˜Š", "Happy to help!"]],

    [r"(bye|exit|quit)",
     ["Goodbye! Have a great day ðŸ‘‹", "See you soon! ðŸ˜Š"]],

    [r"(.*)",
     ["Sorry, I didn't understand that ðŸ¤”. Can you rephrase?",
      "I'm still learning. Please try asking something else."]]
]

chatbot = Chat(pairs, reflections)

print("ðŸ¤– AI Chatbot (type 'bye' to stop)")

while True:
    user_input = input("You: ")
    if user_input.lower() in ["bye", "exit", "quit"]:
        print("Bot: Goodbye! ðŸ‘‹")
        break
    response = chatbot.respond(user_input)
    print("Bot:", response)
```
